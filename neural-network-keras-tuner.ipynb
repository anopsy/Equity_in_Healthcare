# I removed missing values, standarized and scaled my numeric values and one-hot encoded the categorical features. I'm now ready to tune my model. I tune the number of layers, the dropout rate and activation function 

```python
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Dense(units=hp.Int('units_input', min_value=32, max_value=512, step=32), 
                           input_dim=X_train.shape[1], 
                           activation='relu'))  # First layer with input_dim=83
    for i in range(hp.Int('num_layers', 1, 3)):
        model.add(layers.Dense(units=hp.Int('units_' + str(i), 32, 512, 32),
                               activation=hp.Choice('activation_' + str(i), ['relu', 'tanh'])))
        if hp.Choice('dropout_' + str(i), [0, 1]):
            model.add(layers.Dropout(rate=hp.Float('dropout_rate_' + str(i), 0.1, 0.5, step=0.1)))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(optimizer=hp.Choice('optimizer', ['adam', 'sgd']),
                  loss='binary_crossentropy',
                  metrics=[keras.metrics.AUC(name='auc')])
    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective=keras_tuner.Objective("val_auc", direction="max"),
    max_trials=100,
    executions_per_trial=3,
    directory='.',
    project_name='90Diag')
```

```python
tuner.search_space_summary()
```

```python
tuner.search(X_train, y_train,
          epochs=10,
          validation_data=(X_test, y_test),
          batch_size=128)
```

```python
# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]

# Make predictions on the test data
predictions = best_model.predict(X_test)

print(predictions)
```
